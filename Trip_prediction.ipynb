```python
def  my_model(data, n, index):
    i = 0
    trip_id = 0
    transfer_id = 0
    trips = pd.DataFrame([], columns=columns)
    start = data.loc[index]     #start would be used in the final trip
    end = data.loc[index+n-1]
    #start_data = [trip_id, start['NUMR_CART'], start['Lib_Arret'], '', start['HEUR_VALD'], '', start['COD_TLB_RES'], 'd' ]
   # trips.loc[0] = start_data
    trip_start_time = start['HEUR_VALD']   
    while i+1 < n:
        first = data.loc[index+i]
        second = data.loc[index+i+1]
        first_data = [trip_id, first['NUMR_CART'], first['Lib_Arret'], '', first['HEUR_VALD'], '', first['COD_TLB_RES'], 'd' ]
        trips.loc[transfer_id] = first_data
        trips.loc[transfer_id, 'D'] = second['Lib_Arret']
        trips.loc[transfer_id, 'Dt'] = second['HEUR_VALD']
        if second['COD_NATR_VALD'] in sortie:
            trips.loc[transfer_id, 'unkonwn'] = 'no'
            transfer_id = transfer_id + 1 
            i = i + 2
        else:
            trips.loc[transfer_id, 'unkonwn'] = 'd'
            transfer_id = transfer_id + 1
            i = i + 1
            if (second['HEUR_VALD'] - trip_start_time).seconds > T:
                trip_start_time = second['HEUR_VALD']
                trip_id = trip_id + 1
    if (i+1) == n:
        if (end['HEUR_VALD'] -  trip_start_time).seconds > T:
            trip_id = trip_id + 1
        last_data = [trip_id, end['NUMR_CART'], end['Lib_Arret'], '', end['HEUR_VALD'], '', end['COD_TLB_RES'], 'd' ]
        trips.loc[transfer_id] = last_data
        trips.loc[transfer_id, 'trip_id'] = trip_id
        trips.loc[transfer_id, 'D'] = start['Lib_Arret']
        trips.loc[transfer_id, 'unkonwn'] = 'dh'
    return trips
        
                
        
```
