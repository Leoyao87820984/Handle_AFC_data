```python
import pandas as pd
import numpy as np
from IPython.display import Image
from datetime import datetime

pd.options.display.max_rows = 100

def get_counts(anything):
    counts = {}
    for a in anything:
        if a in counts:
            counts[a] = counts[a] + 1
        else:
            counts[a] = 1
    return counts 

```
