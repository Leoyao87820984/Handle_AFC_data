```python
paths = []
for i in range(1,28):
    paths.append('F:/Stage/data/AFC/parts/part_%d.txt'%i)
def get_counts(anything):
    counts = {}
    for a in anything:
        if a in counts:
            counts[a] = counts[a] + 1
        else:
            counts[a] = 1
    return counts
```
